### Statistics Advance Part 1

1. **What is a random variable in probability theory?**  
   A function that assigns numerical values to outcomes of a random experiment.  

2. **What are the types of random variables?**  
   - **Discrete:** Takes countable values (e.g., 0, 1, 2...).  
   - **Continuous:** Takes infinite values within a range.  

3. **What is the difference between discrete and continuous distributions?**  
   - **Discrete:** Defined at specific values.  
   - **Continuous:** Defined over an interval, requiring integration to find probabilities.  

4. **What are probability distribution functions (PDF)?**  
   A function that describes the likelihood of different outcomes for a random variable.  

5. **How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?**  
   - **PDF:** Gives probability at a specific value (for discrete) or density (for continuous).  
   - **CDF:** Cumulative probability up to a certain point.  

6. **What is a discrete uniform distribution?**  
   A distribution where all outcomes have an equal probability (e.g., rolling a fair die).  

7. **What are the key properties of a Bernoulli distribution?**  
   - Only two outcomes: Success (1) or Failure (0).  
   - Probability of success = p, failure = (1 - p).  

8. **What is the binomial distribution, and how is it used in probability?**  
   Models the number of successes in a fixed number of independent Bernoulli trials.  

9. **What is the Poisson distribution and where is it applied?**  
   Models the probability of a given number of events occurring in a fixed interval, used for rare events.  

10. **What is a continuous uniform distribution?**  
    A distribution where all values in a range are equally likely.  

11. **What are the characteristics of a normal distribution?**  
    - Bell-shaped, symmetric around the mean.  
    - Mean = Median = Mode.  
    - Defined by mean (μ) and standard deviation (σ).  

12. **What is the standard normal distribution, and why is it important?**  
    A normal distribution with mean = 0 and standard deviation = 1; used for standardization.  

13. **What is the Central Limit Theorem (CLT), and why is it critical in statistics?**  
    The distribution of sample means approaches normality as sample size increases, regardless of the original distribution.  

14. **How does the Central Limit Theorem relate to the normal distribution?**  
    It allows normal approximation of non-normal distributions when sample size is large.  

15. **What is the application of Z statistics in hypothesis testing?**  
    Used to determine how many standard deviations a value is from the mean.  

16. **How do you calculate a Z-score, and what does it represent?**  
    - **Formula:** \( Z = \frac{(X - \mu)}{\sigma} \)  
    - Represents how far a value is from the mean in terms of standard deviations.  

17. **What are point estimates and interval estimates in statistics?**  
    - **Point Estimate:** Single best guess of a parameter.  
    - **Interval Estimate:** A range within which the parameter likely falls (e.g., confidence interval).  

18. **What is the significance of confidence intervals in statistical analysis?**  
    Provides a range of values that likely contain the true population parameter with a given confidence level.  

19. **What is the relationship between a Z-score and a confidence interval?**  
    Z-scores determine critical values for confidence intervals.  

20. **How are Z-scores used to compare different distributions?**  
    They standardize different distributions, allowing comparison on the same scale.  

21. **What are the assumptions for applying the Central Limit Theorem?**  
    - Sample size is large (usually n ≥ 30).  
    - Independent random sampling.  
    - Finite variance in population.  

22. **What is the concept of expected value in a probability distribution?**  
    The weighted average of all possible values of a random variable.  

23. **How does a probability distribution relate to the expected outcome of a random variable?**  
    The expected value provides the long-run average outcome of a random variable based on its probability distribution.

```python
# 3. Calculate the PDF of a Bernoulli Distribution

from scipy.stats import bernoulli

def bernoulli_pdf(p, x):
    return bernoulli.pmf(x, p)

p = 0.5  # Probability of success
x_values = [0, 1]

for x in x_values:
    print(f"P(X={x}) =", bernoulli_pdf(p, x))
```

```python
# 9. Calculate Z-Scores and Plot Them

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def calculate_z_scores(data):
    return (data - np.mean(data)) / np.std(data)

data = np.random.normal(50, 10, 1000)  # Example dataset
z_scores = calculate_z_scores(data)

plt.hist(z_scores, bins=30, density=True, alpha=0.6, color='blue', edgecolor='black')
plt.xlabel("Z-score")
plt.ylabel("Density")
plt.title("Z-score Distribution")
plt.show()
```

```python
# 10. Implement the Central Limit Theorem (CLT) using a Non-Normal Distribution

import numpy as np
import matplotlib.pyplot as plt

# Generate a non-normal distribution (exponential)
population = np.random.exponential(scale=2, size=10000)

# Draw sample means (CLT effect)
sample_means = [np.mean(np.random.choice(population, 30)) for _ in range(1000)]

plt.hist(sample_means, bins=30, density=True, alpha=0.6, color='red', edgecolor='black')
plt.xlabel("Sample Mean")
plt.ylabel("Density")
plt.title("Central Limit Theorem in Action (Sample Size = 30)")
plt.show()
```

```python
# 18. Calculate the Z-Score for a Given Data Point and Compare to a Standard Normal Distribution

import scipy.stats as stats

def calculate_z_score(value, mean, std):
    return (value - mean) / std

data_point = 75
mean, std = 50, 10
z_score = calculate_z_score(data_point, mean, std)

print("Z-Score:", z_score)
print("Probability from standard normal distribution:", stats.norm.cdf(z_score))
```

```python
# 19. Implement Hypothesis Testing Using Z-Statistics

from scipy.stats import norm

sample_mean = 52
pop_mean = 50
std_dev = 10
n = 30

z_stat = (sample_mean - pop_mean) / (std_dev / np.sqrt(n))
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

print("Z-Statistic:", z_stat)
print("P-Value:", p_value)

if p_value < 0.05:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

```python
# 20. Create a Confidence Interval for a Dataset

import numpy as np
import scipy.stats as stats

data = np.random.normal(50, 10, 100)
mean = np.mean(data)
std_err = stats.sem(data)
confidence = 0.95
margin = stats.t.ppf((1 + confidence) / 2, len(data)-1) * std_err

conf_interval = (mean - margin, mean + margin)
print("95% Confidence Interval:", conf_interval)
```

```python
# 21. Generate Data from Normal Distribution and Calculate Confidence Interval

data = np.random.normal(100, 15, 100)
mean = np.mean(data)
std_err = stats.sem(data)

confidence_interval = stats.norm.interval(0.95, loc=mean, scale=std_err)
print("95% Confidence Interval:", confidence_interval)
```
